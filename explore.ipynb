```python
import pandas as pd
#import datetime as dt
```

```python
#https://www150.statcan.gc.ca/t1/tbl1/en/tv.action?pid=1410029301

# unemployment data  by province from 2001
unemp = pd.read_csv('1410029301_databaseLoadingData_unemp2001.csv')
unemp
```

```python
unemp = unemp.loc[unemp.GEO == 'Canada']
unemp.REF_DATE = pd.to_datetime(unemp.REF_DATE)
```

```python
unemp.VALUE.plot()
```

```python
#https://health-infobase.canada.ca/covid-19/epidemiological-summary-covid-19-cases.html

# total covid cases Canada
covid = pd.read_csv('covid19-download.csv')
covid
```

```python
covid = covid.loc[covid.prname == 'Canada']
covid.date = pd.to_datetime(covid.date)
```

```python
covid['delta'] = covid.numconf.shift(periods=1)
covid.delta =  covid.numconf - covid.delta 
covid.delta.iloc[0] = 4
```

```python
#cov.numconf.plot()
covid.delta.plot()
```
